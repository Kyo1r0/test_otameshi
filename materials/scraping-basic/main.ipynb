```python
from bs4 import BeautifulSoup
import requests
```

```python
# 取得したいURLを指定
url = "https://qiita.com/Chanmoro/items/db51658b073acddea4ac"

# URLからHTMLを取得
# requests.get()でHTTPリクエストを送り、responseオブジェクトを取得
# これはbeautifulsoup4ではなく，単にhttpリクエストを送ってhtmlを取得するだけ
response = requests.get(url)

# BeautifulSoupでHTMLを解析してオブジェクトを作成
# これはbeautifulsoup4でhtmlを解析してhtmlを今後使いやすい形に変換する
soup = BeautifulSoup(response.text, "html.parser")

# 結果を表示
print(soup.prettify())
```

```python
# title タグの文字列を取得する
title = soup.find("title").get_text()
print(title)
```

```python
# hタグを取得する
titles = soup.find_all("h1")
print("titles:", titles)

for title in titles:
    print("title:", title.get_text().strip())
```

```python
# aタグのhref属性を取得する
links = soup.find_all("a")
print("links:", links)

for link in links:
    print(link["href"])
```

```python
# tableタグの中身を取得する
tables = soup.find_all("table")
print("tables:", tables)

for table in tables:
    print(table)
```

```python
# ログインして情報を取得する
# 今回はログイン情報をファイルから読み込む
# 普段やる場合は、ユーザーに入力してもらうなどする
# もしくは環境変数から取得する

with open("password.txt") as f:
    lines = f.readlines()
    username = lines[0].strip()
    password = lines[1].strip()
url = "https://wiki.mma.club.uec.ac.jp/Chofusai2024/Discussion?action=raw"
session = requests.session()
login_info = {
    "name": username,
    "password": password,
    "login": "ログイン",
}
res = session.post(url, data=login_info)
soup = BeautifulSoup(res.content, "html.parser")
print(soup.prettify())
```
